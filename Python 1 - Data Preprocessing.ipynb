## MENGATASI NILAI NULL

```python
df.isna().sum()
```

```python
print(round(100*(df.isnull().sum()/len(df.index)),2))
```

```python
# Cek isi Agent
df[['agent']][df.agent > 0].sample(10)
```

```python
# Cek Describe Agent
df[['agent']].describe()
```

```python
## Mengganti nilai null pada agent dengan median karena data memiliki distribusi skewed
df['agent'].fillna(df['agent'].quantile(0.5), inplace=True)
```

```python
## Drop kolom company karena data tidak bida digunakan (memiliki nilai null sebanyak 94%)
df=df.drop(['company'],axis=1)
```

```python
## Menghapus baris yang berisi nilai Null pada kolom city karena nilai null hanya 0.4%
df = df.dropna(axis = 0)
```

```python
df.isna().sum()
```

## Mengganti value yang tidak sesuai

```python
df[['meal']].value_counts()
```

```python
# Mengganti nilai Undifined dengan Break Fast (mengganti dengan nilai yang sering muncul)
df.replace('Undefined', 'Breakfast', inplace=True)
```

## MEMBUANG DATA TIDAK PENTING

```python
zero_adults = len(df[['adults', 'children', 'babies']][df.adults == 0])
zero_children = len(df[['adults', 'children', 'babies']][df.children == 0])
zero_babies = len(df[['adults', 'children', 'babies']][df.babies == 0])

print(zero_adults)
print(zero_children)
print(zero_babies)
```

```python
print('Presentase Nilai 0 pada kolom adulst sebesar', round(100*(zero_adults/len(df)),2), '%')
print('Presentase Nilai 0 pada kolom children sebesar', round(100*(zero_children/len(df)),2), '%')
print('Presentase Nilai 0 pada kolom babies sebesar', round(100*(zero_babies/len(df)), 2), '%')
```

```python
# Hapus kolom children dan babies karena data tidak bisa digunakan
df=df.drop(['children', 'babies'],axis=1)
```
